```python
import matplotlib.pyplot as plt

# 错误类型和数量
categories = [
    "Lack of Justification or Explanation",
    "Less Precise",
    "Missing Intermediate Steps or Details",
    "Unclear or Inconsistent Notation",
    "Poor Structure or Lack of Formal Presentation",
    "Incomplete Gradient/Hessian Analysis",
    "Lack of Logical Flow",
    "Missing Eigenvalue/Matrix-Related Reasoning"
]
counts = [26, 22, 6, 5, 2, 2, 2, 1]

# 计算百分比标签
total = sum(counts)
percentages = [f"{round(c / total * 100)}%" for c in counts]

# 绘图
fig, ax = plt.subplots(figsize=(12, 6))
wedges, texts = ax.pie(
    counts,
    labels=percentages,            # 饼图上只显示百分比
    startangle=90,
    colors=plt.cm.tab20.colors[:len(categories)],
    labeldistance=1.05,            # 百分比显示在饼图外部
    textprops={'fontsize': 10}
)

# 图例：放在右边，字体更大，去掉边框
ax.legend(
    wedges,
    categories,
    title="Error Categories",
    loc="center left",
    bbox_to_anchor=(1, 0.5),
    fontsize=11,
    title_fontsize=12,
    frameon=False
)

# 设置标题
plt.title("Main Shortcomings of GPT-4o Answers", fontsize=14)

# 保持圆形
ax.axis('equal')

# 布局紧凑
plt.tight_layout()
plt.show()
```
